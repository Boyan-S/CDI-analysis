```python
import numpy as np
import matplotlib.pyplot as plt
import tifffile
from matplotlib.widgets import Slider, Button

plt.rcParams.update({'font.size': 14})
%matplotlib widget
```

```python
# Change path below
path = '/data/Andy/NYZC_rec_0423/A_789/results_phasing/g_0/0/'
rec = np.load(path+'image.npy')

phase = np.angle(rec)
print(np.mean(phase))

# create figure and axis
fig = plt.figure()
ax = fig.add_subplot(111)

# create initial phase map
im = ax.imshow(phase[:,:,0], cmap='viridis', vmin = -np.pi, vmax = np.pi)
plt.colorbar(im, label = 'Phase')

# create slider
axcolor = 'lightgoldenrodyellow'
ax_slider = plt.axes([0.2, 0.0, 0.6, 0.03], facecolor=axcolor)
slider = Slider(ax_slider, 'Z', 0, len(phase[0,0]), valinit=0, valstep = 1)

# update function for slider
def update(val):
    im.set_data(phase[:,:,val])
    fig.canvas.draw_idle()

slider.on_changed(update)
```
